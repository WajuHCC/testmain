```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4],[5,7,9,3],'g^')
plt.axis([0, 10, -10,20])
plt.title('Tytu≈Ç wykresu')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
```

```python
x = np.arange(0, 5, 0.2)
plt.figure(figsize=(10,6))
plt.plot(x,x, 'g--')
plt.plot(x,x**2,'r.-')
plt.plot(x,1/3*x**3, 'k-.')
plt.text(0, 35, "Tekst na wykresie", fontsize=20, color="coral")
arrow = {
    "color" : "blue",
    "linewidth" : 1
}
plt.annotate("Adnotacka na wykresie", [2,5], [0.5,20], arrowprops=arrow)
plt.show()
```
